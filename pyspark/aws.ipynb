```python
import os
import boto3
import numpy as np
import pandas as pd
from io import StringIO, BytesIO
```

```python
def s3_put_csv(bucket_name, key, data):
    csv_buffer = StringIO()
    data.to_csv(csv_buffer)
    s3 = boto3.resource('s3')
    s3.Object(bucket_name, key).put(Body=csv_buffer.getvalue())
    
def s3_delete(bucket_name, key):
    s3 = boto3.resource('s3')
    bucket = s3.Bucket(bucket_name)
    bucket.objects.filter(Prefix = key).delete()
    
def s3_read_csv(bucket_name, key):
    s3 = boto3.resource('s3')
    obj = s3.Object(bucket_name, key)
    body = obj.get()['Body'].read()
    result = pd.read_csv(BytesIO(body))
    return(result)
```

```python
N = 1000
data = pd.DataFrame({'v0': np.random.binomial(10, 0.5, N),
                     'v1': np.random.normal(0, 1, N)})
data.head()
```

```python
s3_put_csv('pyspark-performance', 'data.csv', data)
```

```python
s3_read_csv('pyspark-performance', 'data.csv')
```
